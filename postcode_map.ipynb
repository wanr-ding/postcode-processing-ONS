```python
import pandas as pd
import folium
from IPython.display import display
# import streamlit as st
# from streamlit_folium import st_folium
```

```python
df = pd.read_csv('data/raw/postcode_sample.csv')
```

```python
df.head()
```

```python
# Create base map (centered on UK)
m = folium.Map(location=[54, -2], zoom_start=6)

# Function to determine color based on crime rate
def get_color(crime_rate):
    if crime_rate > 0.1:
        return "red"
    elif crime_rate > 0.07:
        return "orange"
    else:
        return "green"

# Add points
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=6,
        color=get_color(row["crime_rate"]),
        fill=True,
        fill_opacity=0.7,
        popup=(
            f"Postcode: {row['postcode']}<br>"
            f"Crime Rate: {row['crime_rate']}<br>"
            f"Pop Density: {row['pop_density']}"
        )
    ).add_to(m)

# Save map
m.save("data/output/postcode_risk_map.html")

print("Map saved as postcode_risk_map.html")
```
